## Предобработка данных

### Обработка пропусков

Выясним процентное соотношение пропусков в каждом столбце.

```python
df.isna().mean()
```

С каждым из столбцов будем разбираться по отдельности. Также, если необходимо поменять тип данных, то сделаем это сразу в этом пункте, а то, что осталось доделаем в следующих.

#### Столбец *floors_total*

Начнем со столбца *floors_total*. Этот столбец имеет наименьший процент пропусков (~0.4%) среди количественных переменных.
Просмотрим разнообразие значение в этом столбце, чтобы представлять с какими данными имеем дело.

```python
df['floors_total'].value_counts()
```

Данные выглядят вполне реально. Здания с такими количеством этажей не противоречат канонам современной архитектуры.

При таком маленьком проценте пропусков целесообразнее будет удалить эти строки из таблицы, также заменить их проблематично, так как, возможно, получится такая ситуация, когда этаж квартиры получится больше этажности дома, при замене на медиану, даже если ставить логические условия, то это получится с большой натяжкой.

```python
df = df.dropna(subset = ['floors_total'])
```

Изменим тип данных с *float64* на *int64* для более правильного отображения.

```python
df['floors_total'] = df['floors_total'].astype('int64')
```

#### Столбец *ceiling_heght*

Далее работаем со столбцом *ceiling_height*. Процент пропущенных значений равен около 39%. Тут нужна замена медианой или средним значением, но также лучше будет сгрупировать значения столбца *ceiling_height* по значениям *floors_total*.
Сравним значения медианы и среднего арифметического, составив сводную таблицу *height_floor*.

```python
height_floor = df.pivot_table(index ='floors_total', values = 'ceiling_height', aggfunc=['median', 'mean', 'count'])
height_floor.columns = ['median', 'mean', 'count']
height_floor
```

Значения медианы и среднего арифметического отличаются несильно, поэтому заменим пропуски медианой, но возникла другая проблема.
Среди всех объектов недвижимости присутствует только одно здание с высотой в 33 этажа и как раз у этого здания не написана высота потолков.
Предлагаю решить эту проблему, заменив это пустое значению на медиану по всем высотам.
Посмотрим на числовое описание данных в столбце *ceiling_height*.

```python
df['ceiling_height'].describe()
```

Значение медианы расположено на втором квартиле на 5 строке. Произведем замену.

```python
for floor in height_floor.index:
    df.loc[(df['floors_total'].isna()) | ((df['floors_total'] == floor) & (height_floor.loc[floor, 'count'] == 0)), 
             'ceiling_height'] = df['ceiling_height'].describe()[5]
```

Проверим результат.

```python
df[df['floors_total'] == 33]
```

### Изучение параметров

Напишем функцию *research*, которая будет принимать на вход название исследуемой переменной, а на выходе будет выдавать числовое описание данных, диаграмму размаха, значения "усов" и процент выбросов. При написании функции следует учитывать, что если правый "ус" длиннее максимума, то он заканчивается максимумом. То же — для минимума и левого "уса".

```python
def research(variant):
    print('========================================================')
    print('Числовое описание данных')
    print(df[variant].describe())
    
    print('========================================================')
    
    iqr = df[variant].describe()[6] - df[variant].describe()[4]
    
    bottom = df[variant].describe()[4] - 1.5 * iqr
    if bottom < df[variant].describe()[3]: bottom = df[variant].describe()[3]
        
    top = df[variant].describe()[6] + 1.5 * iqr
    if top > df[variant].describe()[7]: top = df[variant].describe()[7]
    
    print('========================================================')
    print('Левый "ус" равен {:.3}'.format(bottom) + ' и правый "ус" равен {:.3}'.format(top))
    
    print('========================================================')
    print('Процент выбросов: {:.1%}'.format(len(df.loc[df[variant] > top]) / len(df)))
    
    print('========================================================')
    print('Диаграмма размаха')
    plt.ylim(0, top*1.2)
    df.boxplot(column = variant)
```

#### Анализ территории

Для анализа напишем функцию *research_spb_center*.

```python
def research_spb_center(variant):
    print('========================================================')
    print('Числовое описание данных')
    print(spb_center[variant].describe())
    
    print('========================================================')
    
    iqr = spb_center[variant].describe()[6] - spb_center[variant].describe()[4]
    
    bottom = spb_center[variant].describe()[4] - 1.5 * iqr
    if bottom < spb_center[variant].describe()[3]: bottom = spb_center[variant].describe()[3]
        
    top = spb_center[variant].describe()[6] + 1.5 * iqr
    if top > spb_center[variant].describe()[7]: top = spb_center[variant].describe()[7]
        
    print('========================================================')
    print('Левый "ус" равен {:.3}'.format(bottom) + ' и правый "ус" равен {:.3}'.format(top))
    
    print('========================================================')
    print('Диаграмма размаха')
    plt.ylim(0, top*1.2)
    spb_center.boxplot(column = variant)
```
